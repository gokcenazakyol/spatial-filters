# Spatial Filters
This exercise will cover the blurring and sharpening filters.

## Import Necessary Libraries and Initial Set-up


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("drive/MyDrive/nature.jpg")
plt.imshow(img);
```

```python
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_RGB);
```

## Smoothing Image

### Averaging Blur

```python
img_blur = cv2.blur(img_RGB, (20,20))
plt.imshow(img_blur);
```

### Median Blur

```python
img_medianBlur = cv2.medianBlur(img_RGB, 5)
plt.imshow(img_medianBlur);
```

### Gaussian Blur

```python
img_gaussianBlur = cv2.GaussianBlur(img_RGB, (5,5), 5)
plt.imshow(img_gaussianBlur);
```

## Sharpening Image

### Laplacian Filter

```python
laplacian = np.array([[0,-1,0],[-1,4,-1],[0,-1,0]])
laplacian
```

```python
img_laplacian = cv2.filter2D(img_RGB, -1, laplacian)
plt.imshow(img_laplacian);
```

### Sobel Filter

```python
img_sobel = cv2.Sobel(img_RGB,-1,1,0)
plt.imshow(img_sobel);
```
